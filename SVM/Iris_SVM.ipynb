```
import numpy as np 
import pandas as pd
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt 
from mlxtend.plotting import plot_decision_regions
```

```
iris=datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['class'] = pd.Series(iris.target)
df.head()
```

```
X=np.array(df.iloc[:,[1,3]])
y=np.array(df.iloc[:,-1])
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```
svcl=SVC(kernel='linear',C=1)
svcl.fit(X_train,y_train)
y_pred=svcl.predict(X_test)
```

```
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```
CM=confusion_matrix(y_test,y_pred)
print(CM)
```

```
CR=classification_report(y_test,y_pred)
print(CR)
```

```
plot_decision_regions(X_test,y_pred,clf=svcl,res=0.02,legend=2)
plt.title('IRIS CLassification by SVM')
plt.show()
```

```
svcp=SVC(kernel='poly',degree=4,C=1,gamma='auto')
svcp.fit(X_train,y_train)
y_pred=svcp.predict(X_test)
```

```
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```
CM=confusion_matrix(y_test,y_pred)
print(CM)
```

```
CR=classification_report(y_test,y_pred)
print(CR)
```

```
plot_decision_regions(X_test,y_pred,clf=svcp,res=0.02,legend=2)
plt.title('IRIS CLassification by SVM')
plt.show()
```
