```
import numpy as np 
import matplotlib.pyplot as plt 
from keras.models import Sequential
from keras.layers import Dense
```

```
np.random.seed(0)
X = np.random.random((1000,100))
print(X.shape)
y = np.random.random((1000,1))
```

```
model = Sequential([
Dense(32,activation ='relu',input_dim =100),
Dense(1, activation ='sigmoid'),
])
# Model Summary
model.summary()
# Compiling the Model
model.compile(loss='mean_squared_error',optimizer='rmsprop', metrics=["mean_absolute_error"])
# Fitting the Model
model.fit(X,y,epochs=1000,batch_size=32,verbose =2)
```

```
predictions = model.predict(X,verbose=1)
print(predictions[:5])
```

```
plt.figure(1)
plt.scatter(y,predictions)
plt.show(block=False)
```

```
plt.figure(2)
Test=plt.scatter(X[:,0],y)
Predict=plt.scatter(X[:,0],predictions)
plt.legend([Predict,Test],["Predicted Data","Real Data"])
plt.show()
```
